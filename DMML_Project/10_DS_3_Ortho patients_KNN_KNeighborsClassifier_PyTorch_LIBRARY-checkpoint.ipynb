```python
# To perform classification using the K-Nearest Neighbors (KNN) algorithm with PyTorch library
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
import torch

# Load the dataset
data = pd.read_csv('column_2C_weka.csv')
```

```python
# Display the first few rows of the dataframe
print(data.head())
```

```python
# Display summary statistics and info to understand the data better
print(data.describe())
```

```python
print(data.info())
```

```python
# checking for categorical column
for col in data.columns:
    print()
    if data[col].dtype == 'object':
        print(f'Name of Column is: {col} and unique values are: {data[col].unique()}')
```

```python
# Function to determine the data type of each column
def categorize_data_types(df):
    data_types = {}
    for column in df.columns:
        unique_values = df[column].dropna().unique()
        num_unique = len(unique_values)
        
        if df[column].dtype == 'object':
            # Check if textual data
            if df[column].str.contains(' ').any():
                data_type = 'Textual'
            # Check if binary categorical data (exactly two unique non-numeric values)
            elif num_unique == 2 and pd.api.types.is_string_dtype(df[column]):
                data_type = 'Binary'
            # Nominal (non-ordered) categories
            else:
                data_type = 'Nominal'
        else:
            # Check if binary categorical data (exactly two unique numeric values)
            if num_unique == 2:
                data_type = 'Binary'
            # Numerical data
            else:
                data_type = 'Numerical'
        
        data_types[column] = data_type
    
    return data_types

# Get and print the data types categorized
categorized_data_types = categorize_data_types(data)
for column, data_type in categorized_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```

```python
# Assuming 'class' is the target variable and it's categorical
X = data.drop('class', axis=1).values
y = data['class'].map({'Abnormal': 1, 'Normal': 0}).values
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
# Convert arrays to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.int64)
y_test = torch.tensor(y_test, dtype=torch.int64)
```

```python
# Implementing KNN in PyTorch
def knn_predict(X_train, y_train, X_test, k=5):
    num_test = X_test.shape[0]
    y_pred = torch.zeros(num_test, dtype=torch.int64)
    
    # Compute pairwise distances between all test points and all train points
    dists = torch.cdist(X_test, X_train)

    for i in range(num_test):
        # Find the k nearest neighbors and their labels
        _, indices = torch.topk(dists[i], k, largest=False)
        nearest_labels = y_train[indices]
        
        # Vote for the most common class
        votes = nearest_labels.bincount()
        y_pred[i] = torch.argmax(votes)
    
    return y_pred
```

```python
# Make predictions
y_pred = knn_predict(X_train, y_train, X_test, k=5)

# Calculate accuracy and F1 score
accuracy = accuracy_score(y_test.numpy(), y_pred.numpy())
f1 = f1_score(y_test.numpy(), y_pred.numpy())

print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='coolwarm', alpha=0.6, edgecolors='k')
plt.title('Actual Classes')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1, 2, 2)
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, cmap='coolwarm', alpha=0.6, edgecolors='k')
plt.title('Predicted Classes')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```
