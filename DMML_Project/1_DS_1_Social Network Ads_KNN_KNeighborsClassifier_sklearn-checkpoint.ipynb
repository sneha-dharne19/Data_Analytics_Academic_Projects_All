```python
# loading the data 
import pandas as pd

# Load the dataset
data = pd.read_csv('Social_Network_Ads.csv')

# Display the first few rows of the dataset
print(data.head())
```

```python
# Function to determine the data type of each column
def categorize_data_types(df):
    data_types = {}
    for column in df.columns:
        unique_values = df[column].dropna().unique()
        num_unique = len(unique_values)
        
        if df[column].dtype == 'object':
            # Check if textual data
            if df[column].str.contains(' ').any():
                data_type = 'Textual'
            # Check if binary categorical data (exactly two unique non-numeric values)
            elif num_unique == 2 and pd.api.types.is_string_dtype(df[column]):
                data_type = 'Binary'
            # Nominal (non-ordered) categories
            else:
                data_type = 'Nominal'
        else:
            # Check if binary categorical data (exactly two unique numeric values)
            if num_unique == 2:
                data_type = 'Binary'
            # Numerical data
            else:
                data_type = 'Numerical'
        
        data_types[column] = data_type
    
    return data_types

# Get and print the data types categorized
categorized_data_types = categorize_data_types(data)
for column, data_type in categorized_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```

```python
# splitting the data into training and testing sets
X = data[['Age', 'EstimatedSalary']]   # features - Independent var
y = data['Purchased'] # target - dependent var
```

```python
from sklearn.model_selection import train_test_split

# Assuming X_scaled and y have been defined previously and scaled accordingly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Number of records in the training data
num_records_train = len(X_train)
print("Number of records in training data:", num_records_train)

# Number of records in the testing data
num_records_test = len(X_test)
print("Number of records in testing data:", num_records_test)
```

```python
from sklearn.preprocessing import StandardScaler

# feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Training the K-NN model on Training set
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train,y_train)
```

```python
# predict test set result
y_pred = classifier.predict(X_test)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.metrics import classification_report, accuracy_score

# Making the Confusion Matrix
cm = confusion_matrix(y_test,y_pred)
print("Confusion matrix:\n", cm)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import numpy as np
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Plotting decision boundaries
X_set, y_set = scaler.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 10, stop=X_set[:, 0].max() + 10, step=1),
                     np.arange(start=X_set[:, 1].min() - 1000, stop=X_set[:, 1].max() + 1000, step=1))
grid_frame = pd.DataFrame(np.column_stack((X1.ravel(), X2.ravel())), columns=['Age', 'EstimatedSalary'])
Z = classifier.predict(scaler.transform(grid_frame)).reshape(X1.shape)

# Visualization
plt.contourf(X1, X2, Z, alpha=0.75, cmap=ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c=ListedColormap(('red', 'green'))(i), label=j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
