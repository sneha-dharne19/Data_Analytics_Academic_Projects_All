```python
#  K-Nearest Neighbors (KNN) regression using PyTorch 
import pandas as pd

# Load the dataset
data = pd.read_csv('insurance.csv')

# Quick look at the data
print(data.head())
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define columns
categorical_columns = ['sex', 'smoker', 'region']
numerical_columns = ['age', 'bmi', 'children']

# Create transformer for preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_columns),
        ('cat', OneHotEncoder(), categorical_columns)
    ])

# Fit and transform the preprocessor on the data
X = preprocessor.fit_transform(data.drop('charges', axis=1))
y = data['charges'].values.reshape(-1, 1)

# Scale the target variable (charges)
y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Split the dataset into the training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.25, random_state=42)

# Convert arrays to PyTorch tensors
import torch

X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32)
```

```python
def knn_regression(X_train, y_train, X_test, k=5):
    num_test = X_test.shape[0]
    y_pred = torch.zeros(num_test, 1)

    # Compute pairwise distances between all test points and all train points
    dists = torch.cdist(X_test, X_train)

    for i in range(num_test):
        # Find the k nearest neighbors and their responses
        _, indices = torch.topk(dists[i], k, largest=False)
        y_pred[i] = y_train[indices].mean()
    
    return y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Predicting on test set
y_pred = knn_regression(X_train, y_train, X_test, k=5)

# Inverse transform predictions
y_pred_inv = y_scaler.inverse_transform(y_pred.detach().numpy())

# Calculate metrics
r2 = r2_score(y_scaler.inverse_transform(y_test.detach().numpy()), y_pred_inv)
mse = mean_squared_error(y_scaler.inverse_transform(y_test.detach().numpy()), y_pred_inv)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_scaler.inverse_transform(y_test.detach().numpy()), y_pred_inv)

print(f'R2 Score: {r2:.2f}')
print(f'MSE: {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
print(f'MAE: {mae:.2f}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_scaler.inverse_transform(y_test.detach().numpy()), y_pred_inv, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('KNN Regression Predictions vs Actual')
plt.show()
```
