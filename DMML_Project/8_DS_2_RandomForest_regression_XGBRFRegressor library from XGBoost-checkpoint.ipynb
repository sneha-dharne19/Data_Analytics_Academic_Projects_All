```python
# RandomForest regression using the XGBRFRegressor library from XGBoost

import pandas as pd

# Load the dataset
data = pd.read_csv('insurance.csv')

# Inspect the data structure
print(data.head())

```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# Encode categorical variables
categorical_features = ['sex', 'smoker', 'region']
data = pd.get_dummies(data, columns=categorical_features, drop_first=True)

# Separate features and target
X = data.drop('charges', axis=1)
y = data['charges']

# Scale the numerical features
scaler = StandardScaler()
X[['age', 'bmi', 'children']] = scaler.fit_transform(X[['age', 'bmi', 'children']])
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into the training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import xgboost as xgb

# Initialize the XGBRFRegressor
model = xgb.XGBRFRegressor(n_estimators=100, max_depth=5, subsample=0.8, colsample_bytree=0.8, random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error

# Predicting the Test set results
y_pred = model.predict(X_test)

# Evaluate results
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)  # Updated to use the new function
mae = mean_absolute_error(y_test, y_pred)

print(f'R² Score: {r2:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')
```

```python
import matplotlib.pyplot as plt
import xgboost as xgb

# Feature importance
xgb.plot_importance(model)
plt.title('Feature Importance')
plt.show()
```
