```python
# Social Networking Ads - KNN classifier using the PyTorch library
# import package

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, accuracy_score
import torch


```

```python
# Load data
data = pd.read_csv('Social_Network_Ads.csv')
# Display the first few rows of the dataset
print(data.head())

```

```python
X = data[['Age', 'EstimatedSalary']].values
y = data['Purchased'].values
```

```python
# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

# Number of records in the training data
num_records_train = len(X_train)
print("Number of records in training data:", num_records_train)

# Number of records in the testing data
num_records_test = len(X_test)
print("Number of records in testing data:", num_records_test)
```

```python
# implement the KNN algorithm using PyTorch
def euclidean_distance(x1, x2):
    return torch.sqrt(torch.sum((x1 - x2) ** 2))

class KNN:
    def __init__(self, k=3):
        self.k = k
    
    def fit(self, X, y):
        self.X_train = torch.tensor(X, dtype=torch.float32)
        self.y_train = torch.tensor(y, dtype=torch.int64)
    
    def predict(self, X):
        X = torch.tensor(X, dtype=torch.float32)
        predictions = [self._predict(x) for x in X]
        return torch.tensor(predictions, dtype=torch.int64)
    
    def _predict(self, x):
        distances = torch.stack([euclidean_distance(x, x_train) for x_train in self.X_train])
        k_indices = torch.argsort(distances)[:self.k]
        k_nearest_labels = self.y_train[k_indices]
        most_common = torch.mode(k_nearest_labels).values.item()
        return most_common

```

```python
# Initialize and train KNN
knn = KNN(k=5)
knn.fit(X_train, y_train)

# Predictions
y_pred = knn.predict(X_test)
```

```python
# evaluating model using accuracy and F1-score

# Calculate accuracy and F1-score
accuracy = accuracy_score(y_test, y_pred.numpy())
f1 = f1_score(y_test, y_pred.numpy())

print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
import matplotlib.pyplot as plt

# Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual')
plt.scatter(range(len(y_pred)), y_pred, color='red', alpha=0.5, label='Predicted')
plt.title('Comparison of Actual and Predicted Values')
plt.xlabel('Sample Index')
plt.ylabel('Classification Outcome')
plt.legend()
plt.show()
```
