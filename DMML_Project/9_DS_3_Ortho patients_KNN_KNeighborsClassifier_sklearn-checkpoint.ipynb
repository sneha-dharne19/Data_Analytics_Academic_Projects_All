```python
# checking for categorical column
for col in data.columns:
    print()
    if data[col].dtype == 'object':
        print(f'Name of Column is: {col} and unique values are: {data[col].unique()}')
```

```python
# Function to determine the data type of each column
def categorize_data_types(df):
    data_types = {}
    for column in df.columns:
        unique_values = df[column].dropna().unique()
        num_unique = len(unique_values)
        
        if df[column].dtype == 'object':
            # Check if textual data
            if df[column].str.contains(' ').any():
                data_type = 'Textual'
            # Check if binary categorical data (exactly two unique non-numeric values)
            elif num_unique == 2 and pd.api.types.is_string_dtype(df[column]):
                data_type = 'Binary'
            # Nominal (non-ordered) categories
            else:
                data_type = 'Nominal'
        else:
            # Check if binary categorical data (exactly two unique numeric values)
            if num_unique == 2:
                data_type = 'Binary'
            # Numerical data
            else:
                data_type = 'Numerical'
        
        data_types[column] = data_type
    
    return data_types

# Get and print the data types categorized
categorized_data_types = categorize_data_types(data)
for column, data_type in categorized_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming the target variable is 'class'
X = data.drop('class', axis=1)
y = data['class']

```

```python
from sklearn.model_selection import train_test_split

# Assuming X_scaled and y have been defined previously and scaled accordingly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Number of records in the training data
num_records_train = len(X_train)
print("Number of records in training data:", num_records_train)

# Number of records in the testing data
num_records_test = len(X_test)
print("Number of records in testing data:", num_records_test)
```

```python
# Encode the target variable if it's categorical
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42)
```

```python
# using the KNeighborsClassifier from scikit-learn.

from sklearn.neighbors import KNeighborsClassifier

# Initialize and train the KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

# Make predictions
y_pred = knn.predict(X_test)

# Calculate accuracy and F1 score
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)  # Adjust the 'average' parameter as needed for multi-class

print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

```python
# Actual vs Predicted plot
plt.figure(figsize=(10, 5))
plt.plot(y_test, label='Actual', marker='o')
plt.plot(y_pred, label='Predicted', linestyle='--', marker='x')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```
