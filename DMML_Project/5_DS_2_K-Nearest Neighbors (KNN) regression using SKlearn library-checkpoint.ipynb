```python
# K-Nearest Neighbors (KNN) regression using SKlearn library
import pandas as pd

# Load the dataset
data = pd.read_csv('insurance.csv')

# Inspect the data structure
print(data.head())

```

```python
print(data.describe())
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encoding categorical variables
label_encoders = {}
for column in ['sex', 'smoker', 'region']:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

```

```python
# Separate features and target
X = data.drop('charges', axis=1) # features - Independent var
y = data['charges']

from sklearn.model_selection import train_test_split

# Assuming X_scaled and y have been defined previously and scaled accordingly
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Number of records in the training data
num_records_train = len(X_train)
print("Number of records in training data:", num_records_train)

# Number of records in the testing data
num_records_test = len(X_test)
print("Number of records in testing data:", num_records_test)
```

```python
# Function to determine the data type of each column
def categorize_data_types(df):
    data_types = {}
    for column in df.columns:
        unique_values = df[column].dropna().unique()
        num_unique = len(unique_values)
        
        if df[column].dtype == 'object':
            # Check if textual data
            if df[column].str.contains(' ').any():
                data_type = 'Textual'
            # Check if binary categorical data (exactly two unique non-numeric values)
            elif num_unique == 2 and pd.api.types.is_string_dtype(df[column]):
                data_type = 'Binary'
            # Nominal (non-ordered) categories
            else:
                data_type = 'Nominal'
        else:
            # Check if binary categorical data (exactly two unique numeric values)
            if num_unique == 2:
                data_type = 'Binary'
            # Numerical data
            else:
                data_type = 'Numerical'
        
        data_types[column] = data_type
    
    return data_types

# Get and print the data types categorized
categorized_data_types = categorize_data_types(data)
for column, data_type in categorized_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```

```python
# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

# Splitting the dataset into the training set and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Initialize KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)

# Fitting KNN to the Training set
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_squared_log_error

# Predicting the Test set results
y_pred = knn.predict(X_test)

# Evaluating results
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_log_error(y_test, y_pred, squared=False)  # Use the root_mean_squared_error function
mae = mean_absolute_error(y_test, y_pred)

print(f'R² Score: {r2:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')
```

```python
import matplotlib.pyplot as plt

# Plot for Actual vs Predicted Values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='blue')
plt.title('Actual vs Predicted Values')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Ideal line for perfect prediction
plt.show()

# Residual Plot
plt.figure(figsize=(10, 6))
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, alpha=0.5, color='red')
plt.title('Residual Plot')
plt.xlabel('Predicted Charges')
plt.ylabel('Residuals')
plt.axhline(y=0, color='black', linestyle='--')
plt.show()
```
