```python
#  RandomForestClassifier using the XGBoost library
import pandas as pd

# Load the dataset
data = pd.read_csv('column_2C_weka.csv')
```

```python
# Display the first few rows of the dataframe
print(data.head())
```

```python
# Display summary statistics and info to understand the data better
print(data.describe())
```

```python
print(data.info())
```

```python
# checking for categorical column
for col in data.columns:
    print()
    if data[col].dtype == 'object':
        print(f'Name of Column is: {col} and unique values are: {data[col].unique()}')
```

```python
# Function to determine the data type of each column
def categorize_data_types(df):
    data_types = {}
    for column in df.columns:
        unique_values = df[column].dropna().unique()
        num_unique = len(unique_values)
        
        if df[column].dtype == 'object':
            # Check if textual data
            if df[column].str.contains(' ').any():
                data_type = 'Textual'
            # Check if binary categorical data (exactly two unique non-numeric values)
            elif num_unique == 2 and pd.api.types.is_string_dtype(df[column]):
                data_type = 'Binary'
            # Nominal (non-ordered) categories
            else:
                data_type = 'Nominal'
        else:
            # Check if binary categorical data (exactly two unique numeric values)
            if num_unique == 2:
                data_type = 'Binary'
            # Numerical data
            else:
                data_type = 'Numerical'
        
        data_types[column] = data_type
    
    return data_types

# Get and print the data types categorized
categorized_data_types = categorize_data_types(data)
for column, data_type in categorized_data_types.items():
    print(f"Column: {column}, Data Type: {data_type}")
```

```python
# Data pre processing
# Assuming that the dataset's target variable is categorical (e.g., class), and other features are numerical:
from sklearn.preprocessing import LabelEncoder

# Encode the categorical target variable
label_encoder = LabelEncoder()
data['class'] = label_encoder.fit_transform(data['class'])

# Separate features and target
X = data.drop('class', axis=1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
import xgboost as xgb

# Initialize the XGBoost classifier
model = xgb.XGBClassifier(
    n_estimators=100,    # Number of trees (like forest size)
    max_depth=5,         # Depth of each tree
    learning_rate=0.1,   # Learning rate
    subsample=0.8,       # Subsampling of the dataset (mimics bagging)
    colsample_bytree=0.8,# Subsampling of features for each tree
    objective='binary:logistic',  # Binary classification
    eval_metric='logloss',
    use_label_encoder=False
)

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

# Predict on the testing data
y_pred = model.predict(X_test)

# Calculate accuracy and F1 score
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print(f'F1 Score: {f1:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Plot feature importance
xgb.plot_importance(model)
plt.title('Feature Importance')
plt.show()

```

```python
# Plot confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
